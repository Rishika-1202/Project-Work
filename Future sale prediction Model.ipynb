```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

using linear regression we will build the model and predict the future sales of the vehicle

```python
da.columns
```

```python
da.head()
```

```python
train = da.drop(['car_ID','symboling','CarName' ,'fueltype','aspiration' , 'doornumber'  ,'carbody','drivewheel','enginelocation'  ,            
 'curbweight','enginetype','cylindernumber','enginesize' ,'fuelsystem' ,'horsepower','peakrpm','citympg','highwaympg','price' ],axis=1)
test = da['price']
```

```python
x_train, x_test, y_train, y_test= train_test_split(train, test, test_size=0.25,random_state=2)
```

```python
model =LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
final=model.predict(x_test)
final
```

```python
final
```

```python
model.score(x_test,y_test)
```
